1. Normal Equation:
θ^=(XTX)−1XTy 
θ^y is the value of θ that minimizes the cost function. is the vector of target values containing y(1) to y(m).

```python
import numpy as np
import matplotlib.pyplot as plt
#generate some random data
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)
#Graph 1: Original Data
plt.figure()
plt.plot(x,y,"b.")
plt.title("Original Data")
plt.show() #The code pauses here until I close this window
# Compute Normal Equation:
# We are adding a column of 1s to X to create X_b (design matrix)
x_b = np.c_[np.ones((100, 1)), x]
theta_best = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)
print("Theta from Normal Equation:")
print(theta_best)
# Generate Predictions
x_new=np.array([[0],[2]])
x_new_b=np.c_[np.ones((2,1)),x_new]
y_predict=x_new_b.dot(theta_best)
# Graph 2: Graph with best fit line
plt.figure()
plt.title("Graph with predicted y")
plt.plot(x,y,"b.",label="data points")
plt.plot(x_new,y_predict,"y-",label="best fit line")
plt.legend()
plt.show()
```

2. Gradient Descent
Gradient Descent is a very generic optimization algorithm capable of finding optimal solutions to a wide range of problems. The general idea of Gradient Descent is to tweak parameters iteratively in order to minimize a cost function.

```python
import numpy as np
import matplotlib.pyplot as plt
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)
plt.figure()
plt.plot(x,y,"b.")
plt.title("Original Data")
plt.show()
x_b = np.c_[np.ones((100, 1)), x]
alpha = 0.1 #learning rate
iterations = 1000 #no. of iterations
m = len(x) #no.of training example
theta = np.random.randn(2,1) #random initialization of parameter
cost_history = [] #store cost value
for i in range(iterations):
  y_predict=x_b.dot(theta)
  error = y_predict - y
  cost = (1/(2*m)) * np.sum(error**2) #cost function
  cost_history.append(cost)
  gradients = (1/m) * x_b.T.dot(error)
  theta = theta - alpha * gradients #update theta
print("Theta from Gradient Descent:")
print(theta)
```

```python
#Plot gradient descent regression line
import numpy as np
import matplotlib.pyplot as plt
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)
plt.figure()
plt.plot(x,y,"b.")
plt.title("Original Data")
plt.show()
x_b = np.c_[np.ones((100, 1)), x]
alpha = 0.1 #learning rate
iterations = 1000 #no. of iterations
m = len(x) #no.of training example
theta = np.random.randn(2,1) #random initialization of parameter
cost_history = [] #store cost value
for i in range(iterations):
  y_predict=x_b.dot(theta)
  error = y_predict - y
  cost = (1/(2*m)) * np.sum(error**2) #cost function
  cost_history.append(cost)
  gradients = (1/m) * x_b.T.dot(error)
  theta = theta - alpha * gradients #update theta
print("Theta from Gradient Descent:")
print(theta)
x_new = np.array([[0],[2]])
x_new_b = np.c_[np.ones((2,1)),x_new]

y_predict_gd = x_new_b.dot(theta)

plt.figure()
plt.title("Gradient Descent Regression Line")

plt.plot(x,y,"b.",label="Data Points")
plt.plot(x_new,y_predict_gd,"r-",label="Gradient Descent")

plt.legend()
plt.show()
```
